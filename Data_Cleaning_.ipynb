```python
#Importing of libraries and importing of titanic dataset
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
path='/content/drive/MyDrive/Dataset/titanic.csv'
tit_df=pd.read_csv(path)
tit_df
```

```python
tit_df.shape #Showing of the number of rows and columns
```

```python
tit_df.describe()
```

```python
tit_df.info()
```

```python
tit_df.isna() #Finding of missing value
```

```python
tit_df.isnull().sum() #Finding the sum of missing value
```

```python
tit_df.isnull().sum().sum() #Overall count of all missing value in a dataset
```

```python
sns.heatmap(tit_df.isnull(),yticklabels=False) #Visual representation of missing value
```

```python
df=tit_df.dropna() #Removing of missing values
```

```python
df.isnull().sum()
```

```python
df.head() #Showing of the first five rows(After Removed)
```

REMOVING OF THE OUTLIERS
